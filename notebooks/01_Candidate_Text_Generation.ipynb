```python
def build_candidate_text(candidate):

    profile = candidate["profile"]
    signals = candidate["redrob_signals"]

    text = []

    #################################################
    # PROFILE
    #################################################

    current_title = profile["current_title"]

    text.append(f"Current role: {current_title}.")
    text.append(f"Primary profession: {current_title}.")
    text.append(f"Current company: {profile['current_company']}.")
    text.append(f"Industry: {profile['current_industry']}.")
    text.append(f"Location: {profile['location']}.")
    text.append(
        f"Professional experience of {profile['years_of_experience']} years."
    )

    if profile["summary"]:
        text.append("Professional Summary:")
        text.append(profile["summary"])

    #################################################
    # SKILLS
    #################################################

    text.append("Skills:")

    for skill in candidate["skills"]:

        text.append(
            f"{skill['name']} "
            f"{skill['proficiency']} skill "
            f"with {skill['duration_months']} months experience "
            f"and {skill['endorsements']} endorsements."
        )

    #################################################
    # CAREER HISTORY
    #################################################

    text.append("Career History:")

    for job in candidate["career_history"]:

        text.append(
            f"Worked as {job['title']} at {job['company']} "
            f"for {job['duration_months']} months."
        )

        if job["description"]:
            text.append(job["description"])

    #################################################
    # EDUCATION
    #################################################

    text.append("Education:")

    for edu in candidate["education"]:

        text.append(
            f"{edu['degree']} "
            f"in {edu['field_of_study']} "
            f"from {edu['institution']} "
            f"{edu['tier']} institution."
        )

    #################################################
    # CERTIFICATIONS
    #################################################

    if len(candidate["certifications"]) > 0:

        text.append("Certifications:")

        for cert in candidate["certifications"]:

            text.append(
                f"{cert['name']} issued by {cert['issuer']}."
            )

    #################################################
    # LANGUAGES
    #################################################

    text.append("Languages:")

    for lang in candidate["languages"]:

        text.append(
            f"{lang['language']} "
            f"{lang['proficiency']}."
        )

    #################################################
    # BEHAVIORAL SIGNALS
    #################################################

    text.append("Behavioral Signals:")

    if signals["open_to_work_flag"]:
        text.append("Candidate is open to work.")
    else:
        text.append("Candidate is not actively open to work.")

    github_score = signals["github_activity_score"]

    if github_score > 50:
        text.append(
            f"Strong Github activity score of {github_score}."
        )

    elif github_score > 0:
        text.append(
            f"Moderate Github activity score of {github_score}."
        )

    else:
        text.append(
            "No Github activity."
        )

    response_rate = signals["recruiter_response_rate"]

    if response_rate > 0.7:
        text.append(
            f"Excellent recruiter response rate of {response_rate*100:.0f} percent."
        )

    elif response_rate > 0.4:
        text.append(
            f"Moderate recruiter response rate of {response_rate*100:.0f} percent."
        )

    else:
        text.append(
            f"Low recruiter response rate of {response_rate*100:.0f} percent."
        )

    text.append(
        f"Interview completion rate is "
        f"{signals['interview_completion_rate']*100:.0f} percent."
    )

    text.append(
        f"Saved by recruiters "
        f"{signals['saved_by_recruiters_30d']} times."
    )

    notice_period = signals["notice_period_days"]

    if notice_period <= 30:
        text.append(
            f"Short notice period of {notice_period} days."
        )

    elif notice_period <= 60:
        text.append(
            f"Moderate notice period of {notice_period} days."
        )

    else:
        text.append(
            f"Long notice period of {notice_period} days."
        )

    return "\n".join(text)
```

```python
candidate_text = build_candidate_text(candidates[0])

print(candidate_text)
```

```python
candidate_texts = []

for c in tqdm(candidates):

    candidate_texts.append(
        build_candidate_text(c)
    )
```
