```python
import pickle
import numpy as np
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
```

```python
with open(
"/content/drive/MyDrive/Redrob_AI_Hackathon/outputs/top5000_bm25.pkl",
"rb"
) as f:

    top_candidates = pickle.load(f)

print(len(top_candidates))
```

```python
model = SentenceTransformer(
    "BAAI/bge-small-en-v1.5"
)
```

```python
query = """
Senior Machine Learning Engineer

AI Engineer

Search Engineer

Recommendation Systems

RAG

Embeddings

Vector Databases

FAISS

Milvus

Pinecone

BM25

PyTorch

Transformers

Semantic Search

Python

Open to work

Strong recruiter response

Short notice period

5 to 9 years experience
"""
```

```python
query_embedding = model.encode(
    query,
    normalize_embeddings=True
)
```

```python
candidate_texts = [
    x["text"]
    for x in top_candidates
]
```

```python
candidate_embeddings = model.encode(
    candidate_texts,
    batch_size=32,
    show_progress_bar=True,
    normalize_embeddings=True
)
```

```python
similarities = (
    candidate_embeddings @ query_embedding
)
```

```python
top_indices = np.argsort(
    similarities
)[::-1][:1000]
```

```python
top1000_dense = []

for idx in top_indices:

    top1000_dense.append(
        top_candidates[idx]
    )
```

```python
with open(
"/content/drive/MyDrive/Redrob_AI_Hackathon/outputs/top1000_dense.pkl",
"wb"
) as f:

    pickle.dump(
        top1000_dense,
        f
    )
```
