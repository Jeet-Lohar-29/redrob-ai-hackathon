```python
import pickle
from rank_bm25 import BM25Okapi
from tqdm import tqdm
```

```python
with open(
"/content/drive/MyDrive/Redrob_AI_Hackathon/embeddings/candidate_texts.pkl",
"rb"
) as f:

    candidate_texts = pickle.load(f)


with open(
"/content/drive/MyDrive/Redrob_AI_Hackathon/embeddings/candidate_ids.pkl",
"rb"
) as f:

    candidate_ids = pickle.load(f)


print(len(candidate_texts))
```

```python
tokenized_corpus = []

for doc in tqdm(candidate_texts):
    tokenized_corpus.append(
        doc.lower().split()
    )
```

```python
bm25 = BM25Okapi(tokenized_corpus)
```

```python
with open(
"/content/drive/MyDrive/Redrob_AI_Hackathon/models/bm25_model.pkl",
"wb"
) as f:

    pickle.dump(bm25,f)
```

```python
query = """
Machine Learning Engineer

AI Engineer

Search and Retrieval

LLM Practitioner

Python

RAG

BM25

FAISS

Vector Databases

Milvus

Pinecone

Embeddings

LangChain

LLM Fine-tuning

PyTorch

Transformers

Semantic Search

Recommendation Systems

5 to 9 years experience

Open to work

Strong Github activity

High recruiter response rate

Short notice period
"""

query_tokens = query.lower().split()
```

```python
scores = bm25.get_scores(query_tokens)
```

```python
import numpy as np

top_k = 5000

top_indices = np.argsort(scores)[::-1][:top_k]
```

```python
top_candidates = []

for idx in top_indices:

    top_candidates.append({

        "candidate_id":candidate_ids[idx],

        "score":scores[idx],

        "text":candidate_texts[idx]

    })
```

```python
for i in range(5):

    print("="*80)

    print(top_candidates[i]["candidate_id"])

    print()

    print(top_candidates[i]["score"])

    print()

    print(top_candidates[i]["text"][:1500])
```

```python
with open(
"/content/drive/MyDrive/Redrob_AI_Hackathon/outputs/top5000_bm25.pkl",
"wb"
) as f:

    pickle.dump(top_candidates,f)
```
